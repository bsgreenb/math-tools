# Divisibility

## Definition of Divisibility

Let $a$ and $b$ be integers with $a \ne 0$. We say $a$ divides $b$, denoted by $a|b$, if there exists an integer $c$ such that $b=ca$.  Alternate phrasings:
- $b$ is divisible by $a$
- $b$ is a multiple of $a$
- $a$ is a divisor of $b$

Examples: $2|6, -3|9, -5|-10, 4|0, 3 \nmid 7$

Trivial facts:
- $1 | n$ for any integer ($\mathbb{Z}$)
- $0 \nmid n$ for any $n \in \mathbb{Z}$
- $n | 0$ for any $n \in \mathbb{Z}, n \ne 0$

## Divisibility Theorems

Let $a, b, c, x, y$ be integers.

1. If $a|b$ then $a|xb$
2. If $a|b$ and $a|c$, then $a|bx + cy$
3. If $a|b$ then $xa|xb$
4. If $a|b$, then $|a| \le |b|$.  In particular, if $a|b$ and $b|a$, then $a=\pm b$.

## Divisibility Tests

I followed along with [AoPS proofs](https://artofproblemsolving.com/wiki/index.php/Divisibility_rules) which were fun in terms of how they broke down a base 10 number and proved things about it, etc.

### Divisibility by $10^n, 5, 25$

- A number is divisible by $10^n$ iff it has $n 0$s at the end.
- A number is divisible by 5 iff the last digit is 5 or 0.
- A number is divisible by 25 iff the last 2 digits are divisible by 25.

### Divisibility by a power of 2
Let $n$ be a positive integer.  A number is divisible by $2^n$ iff the last $n$ digits are divisible by $2^n$.

### Divisbility by 3 and 9

A number is divisible by 3 or 9 if the sum of its digits is divisible by 3 or 9, _respectively_.

### Divisibility by 11

A number N is divisible by 11 if the **alternating sum** of its digits is divisible by 11.

### More divisibility testing

- A number is divisible by 6 iff it's divisible by 2 and 3.
- A number is divisible by 12 iff it's divisible by 3 and 4.

# GCD and LCM

The **greatest common denominator** (gcd) of two nonzero integers a and b is the greatest positive integer d such that d is a divisor of both a and b.

This definition also applies when one of a and b is zero (or typically, when both are).  In this case:

$\text{gcd}(a, 0) = \text{gcd}(0, a) = |a|$

gcd(0,0) is usually defined as 0 (if it's not left undefined), making stuff like Euclidean Division Algorithm Work well.  [MSE on this convention](https://math.stackexchange.com/questions/1386651/is-this-gcd0-0-0-a-wrong-belief-in-mathematics-or-it-is-true-by-conventi).

The GCD will be the item in the set of common divisors that the others go into.  This also serves as a handy justification for $\text{gcd}(0, 0)$.

GCD of multiple numbers:

$$\text{gcd}(a,b,c)=\text{gcd}(a,\text{gcd}(b,c))$$

## Calculation of GCD

- Prime factorization - an approach that is only really feasible for small numbers.
- Euclidean algorithm - will discuss this more in depth

## Lowest common multiple

Let a and b be nonzero integers.  The **lowest common multiple** (lcm) is the smallest possible integer that is divisible by both a and b.

The lcm is the **lowest common denominator** (lcd) that can be used with fractions.  LCM like GCD is defined for multiple integers. LCM and GCD are related as follows:

$$
\text{lcm}(a,b) = \frac{|ab|}{\text{gcd}(a,b)}
$$

# Prime and Coprime Numbers

We say an integer $p > 1$ is **prime** if its only divisors are 1 and itself.

An integer $n > 1$ which is not prime, is said to be a **composite number**.  Composite numbers take the form $n = ab$, where $a$ and $b$ are in interval $(1,n)$.

We say that two nonzero integers a and b are **coprime** or **relatively prime** if $\text{gcd}(a, b) = 1$.  

Two distinct primes will always be coprime.

## Fermat Numbers

A **fermat number** is a positive integer of the form $F_n = 2^{2n} + 1$ where n is a non-negative integer.  The prime ones are called **Fermat Primes**.

Distinct fermat numbers are coprime.

## Merseinne Numbers

Integers of the form $M_p = 2^p - 1$ where p is a prime, are called **Mersenne Numbers**, and when they're prime they're called **Mersenne Primes**.

If $m > 1$ and $a^m -1$ is prime, then $a=2$ and $m$ is prime. [Can prove this later as an exercize when digging deeper on number theory]

# Euclidean Division
**Euclidean Division** is division with remainder, producing a quotient and a remainder that is smaller than the divisor.  When the operation is to only get the remainder, it's called **modulo**.

A fundamental property is that the quotient and the remainder exist, and are unique under some conditions.

## Division Theorem

**Euclidean division** aka the **division algorithm** is based on the following result called the **Division Theorem** also known as **Euclid's Division Lemma**.

Given two integers $a$ (**dividend**) and $b$ (**divisor**), with $b \ne 0$, there exist unique integers $q$ (**quotient**) and $r$ (**remainder**) such that

$$a=bq + r \quad\text{where } 0 \le r < |b|$$

# Euclidean Algorithm

An efficient method for computing GCD.  

Followed along on [Wiki](https://en.wikipedia.org/wiki/Euclidean_algorithm), including the proof of validity. The proof of validity is 2 parts: proof of common divisor, and 2nd part shows it must be the greatest one.

Note how Euclidean Algorithm depends on the Divsion Theorem from section above.

See Algebra for coverage of Euclidean Algorithm for Polynomials.

```python
# Takes two ints and returns greatest common denominator with Euclidean Algorithm
def gcd(a, b):
    # Max handles negative numbers
    if max(a,-a) < b:
        # Ensure a is gte than b
        return gcd(b, a) 
    if b == 0: # Covers (0,0) as 0
        return max(a, -a)
    
    return gcd(b, a % b)

print(gcd(199, 6673))
```

## Extended Euclidean Algorithm
The [Extended Euclidean Algorithm](https://en.wikipedia.org/wiki/Extended_Euclidean_algorithm) is an extension to the Euclidean algorithm, that not only gives you gcd but also the Bezout Coefficients-- $x$ and $y$ in $ax+by = \text{gcd}(a,b)$.

Differs from the standard Euclidean Algorithm in that you're also building up coefficients along the way (see code below).

This algorithm can be used to solve Linear Diophantine Equations (more on that later).

I followed [proof on Wiki](https://en.wikipedia.org/wiki/Extended_Euclidean_algorithm#Proof).

See Algebra for coverage of polynomial case.



```python
# Takes two ints, and returns their GCD and Bezout Coefficients with Extended Euclidean Algorithm
def extended_euclid(a, b):
    old_r, r = a, b
    old_s, s = 1, 0
    old_t, t = 0, 1
    
    while r != 0:
        quotient = old_r // r
        old_r, r = r, old_r - quotient * r
        old_s, s = s, old_s - quotient * s
        old_t, t = t, old_t - quotient * t
    
    # At the end, old_r is gcd(a,b),
    # and old_s, old_t are the Bezout coefficients for a & b, respectively.
    return old_r, (old_s, old_t)

# Test
num_1 = 21
num_2 = 32
gcd_val, (x, y) = extended_euclid(num_1, num_2)
print(f"({num_1} * {x}) + ({num_2} * {y}) = {gcd_val}")

```

# Bezout's Identity

**Bezout's Identity** - Let $a$ and $b$ be integers with gcd $d$.  Then there exist integers $x$ and $y$ such that $ax + by = d$.  

The integers $x$ and $y$ are called **Bezout Coefficents**, and they are not unique. Moreover, the integers of the form $az+bt$ are exactly the multiples of $d$.  For this to work we must take $\text{gcd}(0,0) = 0$. 

A pair of Bezout coefficients can be computed by the extended Euclidean algorithm above. 
Many other theorems in number theory, such as [Euclid's Lemma](https://en.wikipedia.org/wiki/Euclid%27s_lemma), result from Bezout's Identity.

## Structure of Bezout Solutions

If $a$ and $b$ are both non-zero, and one pair of bezout coefficients has been completed, e.g. using extended Euclidean Algorithm, all pairs can be represented as:

$$(x-k\frac{b}{d}, y + k\frac{a}{d}$$

where $k$ is an arbitrary integer, $d$ is the gcd, and the fractions will simplify to integers.

If $a$ and $b$ are both nonzero, then exactly two of these pairs of pairs of BÃ©zout coefficients satisfy

$$
|x| \leq\left|\frac{b}{d}\right| \quad \text { and } \quad|y| \leq\left|\frac{a}{d}\right|
$$

And equality may only occur if one of a or b divides the other (this follows from the Division Theorem).

# Modular Arithmetic

## Congruence

Given an integer $n > 1$, called a **modulus**, two integers are said to be **congruent** modulo $n$ if $n$ is a divisor of their difference, i.e. if there is an integer $k$ such that $a-b = kn$.  Congruence means their difference is a multiple.

$$a \equiv b \pmod n$$

Congruence modulo is an **equivalence relation** compatible with addition, subtraction, and multiplication:

- Reflexivity: $a \equiv a \pmod n$
- Symmetry: $a \equiv b \pmod n$ if $b \equiv a \pmod n$
- Transitivity: if $a \equiv b \pmod n$  and $b \equiv c \pmod n$ then $a \equiv c \pmod n$

If $a_{1} \equiv b_{1}(\bmod n)$ and $a_{2} \equiv b_{2}(\bmod n)$, or if $a \equiv b(\bmod n)$, then:
- $a+k \equiv b+k(\bmod n)$ for any integer $k$ (compatibility with translation)
- $k a \equiv k b(\bmod n)$ for any integer $k$ (compatibility with scaling)
- $a_{1}+a_{2} \equiv b_{1}+b_{2}(\bmod n)$ (compatibility with addition)
- $a_{1}-a_{2} \equiv b_{1}-b_{2}(\bmod n)$ (compatibility with subtraction)
- $a_{1} a_{2} \equiv b_{1} b_{2}(\bmod n)$ (compatibility with multiplication)
- $a^{k} \equiv b^{k}(\bmod n)$ for any non-negative integer $k$ (compatibility with exponentiation)
- $p(a) \equiv p(b)(\bmod n)$, for any polynomial $p(x)$ with integer coefficients (compatibility with polynomial evaluation)

For cancellation of common terms, we have the following rules:
- If $a+k \equiv b+k(\bmod n)$, where $k$ is any integer, then $a \equiv b(\bmod n)$
- If $k a \equiv k b(\bmod n)$ and $k$ is coprime with $n$, then $a \equiv b(\bmod n)$
- If $k a \equiv k b(\bmod k n)$, then $a \equiv b(\bmod n)$

# Diophantine Equations

A polynomial equation, usually involving to or more unknowns, such that the only solutions of interest are the integer ones.  

Diophantine Problems are systems of equations, and there's a whole field of Diophantine geometry.  Individual puzzles were solved over the years, but the general theories of Diophantine equations were an achievement of the 20th century.


## Linear Diophantine Equations

A **linear Diophantine equation** equates to a constant the sum of two or more monomials, each of degree one.  Takes the form

$$ax + by = c$$

where $a$, $b$, and $c$ are given.


## Homogenous Diophantine equations

Defined by a [homogenous polynomial](https://en.wikipedia.org/wiki/Homogeneous_polynomial). A typical such equation is **Fermat's Last Theorem**

$x^d + y^d - z^d = 0$ aka $x^d + y^d = z^d$

Solving these is pretty advanced.  "Rational points on a hypersurface", etc.  Some of them can be solved with my current knowledge though.

## Exponential Disophantine Equations

**Exponential diophantine equation** is one in which unknowns can appear as exponents.
