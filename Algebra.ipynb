# Inequalities

You can apply a **monotonically** increasing function to both sides without changing the inequality. If you apply a monotonically decreasing function to both sides you need to reverse the inequality.

**Multiplication/Division**

If $a \leq b$ and $ c > 0: ac\leq bc$

If $a \leq b$ and $ c < 0: ac\geq bc$

**Additive Inverse**

If $a \leq b$, then $-a \geq -b$

**Multiplicative Inverse**
If $0 < a \leq b$ then $\frac{1}{a} \geq \frac{1}{b} > 0$.

If $a \leq b < 0$ then $ 0 > \frac{1}{a} \geq \frac{1}{b} $.

If $a < 0 < b$ then $\frac{1}{a} < 0 < \frac{1}{b}$.

# Log rules

Will prove each one in terms of previous Exponent and Log rules.

**Log Product Rule**: $$ \log_{b}(MN) = \log_{b}(M) + \log_{b}(N) $$
Proof:

$$
\begin{align*}
& x = \log_{b}(MN) \\
& y =  \log_{b}(M) \\ 
& z = \log_{b}(N) \\
& b^y = M \\
& b^z = N \\
& b^x = MN = b^{y}b^{z} = b^{y + z} \quad\text{Multiplying Exponents} \\
& x = y + z \\
& \log_{b}(MN) = \log_{b}(M) + \log_{b}(N)
\end{align*}
$$

**Log Quotient Rule**: $$ \log_{b}(M/N) = \log_{b}(M) - \log_{b}(N) $$
Proof:

$$
\begin{align*}
& x = \log_{b}(M/N) \\
& y =  \log_{b}(M) \\ 
& z = \log_{b}(N) \\
& b^y = M \\
& b^z = N \\
& b^x = M/N = \frac{b^y}{b^z} = b^{y - z} \quad\text{Dividing Exponents} \\
& x = y - z \\
& \log_{b}(M/N) = \log_{b}(M) - \log_{b}(N)
\end{align*}
$$

**Log Power / Constant Product Rule**: $$ \log_{b}(M^p) = p\log_{b}(M) $$
Proof:

$$
\begin{align*}
& x = \log_{b}(M^p) \\
& y = \log_{b}(M) \\
& b^y = M \\
& b^x = M^p \\
& b^x = (b^y)^p \\
& b^x = b^{yp} \quad\text{Power of a Power} \\
& x = py \\
& \log_{b}(M^p) = p\log_{b}(M) \\
\end{align*}
$$

**Change of Base Rule**: $$ \log_{a}(x) = \frac{\log_{b}(x)}{\log_{b}(a)} $$
Proof:

$$
\begin{align*}
& a^y = x \\
& \log_{b}(a^y) = \log_{b}(x) \\
& y\log_{b}(a) = \log_{b}(x) \quad\text{Log Power Rule} \\
& y = \frac{\log_{b}(x)}{\log_{b}(a)} \\
& \log_{a}(x) = \frac{\log_{b}(x)}{\log_{b}(a)} 
\end{align*}
$$

# Polynomials

## Defining Terms

*Polynomial* One of those equations with coefficients, variables, and non negative exponentiation combined with addition/multiplication/etc.

*Quadratic* functions are polynomials of degree 2.

*Parabola* - graph of a quadratic.  They have those open face shapes and point up or down. 

One definition of a parabola is that it's the set of points which are equidistant from a Focus (a point) and a Directrix (Line).

Forms of equations for a parabola with examples:

- *Standard Form* - $ 3x^2 + 12x - 15 $
- *Factored Form* - $ 3(x-1)(x+5) $
- *Vertex Form* - $ 3(x+2)^2 - 27 $

Abstraction of vertex form we learned in Geometry lesson:
$$ y = a(x-h)^2 + k $$
$$ x = a(y-k)^2 + h $$

Where $ (h, k) $ is the *Vertex* of the parabola. That Vertex is either the top or the bottom.

Aha! Just noticed the Vertex and the Root (derived from factored form) could be same, if there's no constant term.

*Monic* quadratics are expressions where the leading coefficient ($a$) is equal to $1$.

## Fundamental Theorem of Algebra

Any every polynomial equation of degree $ n $ with complex number coefficients has $ n $ roots, or solutions, in the complex numbers.

Explanation of why roots can count twice, such as in $x^2$  which has a repeating 0 root. https://math.stackexchange.com/a/1131546/49487

The FTA guarantees reduction to linear and quadratic factorings of every polynomial with real coefficients.

## Completing the Square

Completing the square works on any quadratic, and is actually the proof basis of the Quadratic Formula.

Make sure to divide out coefficient $a$ before completing the square.  Then just add a constant factor thats the square of half coefficient $b$ to do your squaring proper.

Completing the square works cus it lets you take a square root of both sides to eliminate the 2nd degree factor.

Make sure that when you complete the square and do the final square root step, that you remember this produces a $ \pm $ value.

## Quadratic Formula

$ x=\frac{-b\pm\sqrt{b^2-4ac}}{2a} $

The discriminant is the part of the quadratic formula underneath the square root symbol: $ b^2-4ac $. The discriminant tells us whether there are two solutions, one solution, or no solutions.

* Positive Discriminant - $ b^2 - 4ac > 0 $ - 2 real roots
* Zero Discriminant - $ b^2 - 4ac = 0 $ - 1 repeated real 0 root.
* Negative Discriminant - $ b^2 - 4ac < 0 $ - 2 imaginary roots.

### Proof of Quadratic Formula using Completing the Square

$$
\begin{align*}
ax^2 + bx + c = 0 \\
x^2 + \frac{bx}{a} + \frac{c}{a} = 0 \\
x^2 + \frac{bx}{a} + \frac{c}{a} + ((\frac{b}{2a})^2 - \frac{c}{a}) = ((\frac{b}{2a})^2 - \frac{c}{a}) \\
x^2 + \frac{bx}{a} + (\frac{b}{2a})^2  = (\frac{b}{2a})^2 - \frac{c}{a} \\
(x + \frac{b}{2a})^2 = (\frac{b}{2a})^2 - \frac{c}{a} \quad\text{Complete the Square} \\
(x + \frac{b}{2a})^2 = \frac{b^2}{4a^2} - \frac{c}{a} \\
(x + \frac{b}{2a})^2 = \frac{b^2 - 4ac}{4a^2} \\
x + \frac{b}{2a} = \pm\sqrt{\frac{b^2 - 4ac}{4a^2}} \\
x + \frac{b}{2a} = \frac{\pm\sqrt{b^2 - 4ac}}{2a} \\
x = \frac{-b\pm\sqrt{b^2 - 4ac - b}}{2a}
\end{align*}
$$

## Binomial Theorem

$$ (a+b)^n = \sum_{k=0}^{n}\binom{n}{k}a^{n-k}b^k $$

We know that this is true for $n = 1$, so we can prove it by induction for all positive numbers. So we seek to prove that

$$ 
\begin{align}
& (a+b)^{n+1} = \sum_{k=0}^{n+1}\binom{n+1}{k}a^{n+1-k}b^k \\
& (a + b)(a+b)^n \\
& (a + b)\sum_{k=0}^{n}\binom{n}{k}a^{n-k}b^k \quad\text {Induction} \\
& \left(\sum_{k=0}^{n}\binom{n}{k}a^{n-k + 1}b^k\right) + \left(\sum_{k=0}^{n}\binom{n}{k}a^{n-k}b^{k+1}\right) \\
& \left(\binom{n}{0}a^{n+1}b^0 + \sum_{k=1}^{n}\binom{n}{k}a^{n-k + 1}b^k\right) + \left(\sum_{k=1}^{n+1}\binom{n}{k-1}a^{n-k+1}b^{k}\right) \\
& \left(a^{n+1} + \sum_{k=1}^{n}\binom{n}{k}a^{n-k + 1}b^k\right) + \left(\binom{n}{n}a^{0}b^{n+1} + \sum_{k=1}^{n}\binom{n}{k-1}a^{n-k+1}b^{k}\right) \\
& a^{n+1} + b^{n+1} + \sum_{k=1}^{n}\binom{n}{k}a^{n-k + 1}b^k + \sum_{k=1}^{n}\binom{n}{k-1}a^{n-k+1}b^{k} \\
& a^{n+1} + b^{n+1} + \sum_{k=1}^{n}\binom{n+1}{k}a^{n-k + 1}b^k \quad\text {Pascal's Identity} \\
& \binom{n+1}{0} a^{n-0+1}b^0 + \binom{n+1}{n+1}a^{n-(n+1)+1}b^{n+1} + \sum_{k=1}^{n}\binom{n+1}{k}a^{n-k + 1}b^k \\
& \sum_{k=0}^{n+1}\binom{n+1}{k}a^{n-k + 1}b^k \\
\end{align}
$$ 

## Polynomial Division

The [Division Theorem](https://en.wikipedia.org/wiki/Euclidean_division#Division_theorem) (see Numbers notebook) applies to polynomials with the difference being that degree of the polynomial decreases at each stage rather than the size of the number. 

## Polynomial GCD

The GCD of two polynomials is a polynomial of the highest possible degree that is a factor of both the two polynomials.  

The similarity between the integer GCD and the polynomial GCD allows extending to univariate polynomials all the properties that may be deduced from the Euclidean algorithm and Euclidean division, e.g.

$$\text{gcd}(p,q) = \text{gcd}(q, p+rq) \quad\text{for any polynomial r}$$

See Numbers notebook for overview of the integer rules in depth which all have parallels here.

## Irreducible and Coprime Polynomials
Polynomials with a single variable $x$ (**univariate polynomials**) can be added, multiplied, and factored into **irreducible polynomials**, which are analogs of the prime numbers for integers.

One says $p$ and $q$ are **coprime polynomials** if their GCD is $1$.

## Euclid's Algorithm: Polynomials

Use to find Polynomial GCD. 

Whereas with integers, $r_i$ would decrease at each stage, with polynomials, the degree of the polynomial decreases at each stage.  More specifically, if $b(x) \ne 0$ (in which case GCD is $a(x)$):

The last non-zero remainder (made monic if necessary) is the GCD of the two polynomials. 

The procedure is similar to integers.. at each step k, a quotient polynomial $q_k(x)$ and remainder polynomial $f(x)$ are identified to satisfy the recursive equation:

$$r_{k-2}(x) = q_k(x)r_{k-1}(x) + r_k(x)$$

## Bezout's Identity for Polynomials / Extended Euclidean Alg

$\text{gcd}(p(x), q(x)) = a(x)p(x) + b(x)q(x)$

So with extended Euclidean alg on polynomials, we can actually get the polynomials $p$ and $q$, similar to integer case by keeping track of sequence.

## Polynomial Partial Fraction Decomposition

These are really cool.  Explained on Wiki:

> In symbols, the partial fraction decomposition of a rational fraction of the form $\frac{f(x)}{g(x)}$, where $f$ and $g$ are polynomials, is its expression as
$$
\frac{f(x)}{g(x)}=p(x)+\sum_{j} \frac{f_{j}(x)}{g_{j}(x)}
$$
> where $p(x)$ is a polynomial, and, for each $j$, the denominator $g_{j}(x)$ is a power of an irreducible polynomial, and the numerator $f_{j}(x)$ is a polynomial of a smaller degree than the degree of this irreducible polynomial.

Special cases to be aware of when solving the partial fractions:
- repeating factors
- irreducible quadratics.
I learned/worked through the [Khan Lessons on Partial Fractions](https://www.khanacademy.org/math/algebra-home/alg-rational-expr-eq-func/alg-partial-fraction/v/partial-fraction-expansion-1), which included dealing with those cases.

Thanks to this [MSE answer](https://math.stackexchange.com/a/4362092/49487) for explaining why it works, and its relation to other theorems.

> Given polynomials $\,P(x), Q(x)\,$ with $\,\gcd\big(P(x),Q(x)\big) = 1\,$ i.e. relatively coprime, there exist $\,A(x), B(x)\,$ with $\,\deg A \lt \deg P\,$ and $\,\deg B \lt \deg Q\,$ such that:

$$
\frac{1}{P(x)Q(x)} = \frac{A(x)}{P(x)}+\frac{B(x)}{Q(x)}
$$

# Even vs Odd Functions

*Even functions* can be flipped over the y-axis, *Odd functions* have to be flipped over x and y axis.  

**Even function**:
$$ f(x) = f(-x) $$

**Odd function**
$$ f(x) = -f(-x) $$

Only 0 can be both even and odd (e.g. $f(x) = 0$).  Thing can be neithern (e.g. $(x-2)^2$) depending on whether those equalities hold or not.. Simplest way to check function for even/odd is just to plugin x.  Same value means even, every term being _opposite_ implies odd.
